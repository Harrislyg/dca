```python
spy['EMA20'] = spy['Close'].ewm(span=20, adjust=False).mean()
spy['EMA20_pctchng'] = spy['EMA20'].pct_change(1)*100
spy['EMA20_pctchng'].mean()
```

```python
spy['month'] = pd.DatetimeIndex(spy['Date']).month
spy['next_month'] = spy['month'].shift(-1)
spy.drop(spy.tail(1).index,inplace=True)
```

```python
spy.tail(30)
```

```python
# Dollar cost averaging 50 SMA 2006 - 2019
```

```python
spy = spy[np.isfinite(spy['MA50_pctchng'])]
spy.head()
len(spy.index)
```

```python
number_of_shares = 0

freq_purchase = 0

total_cost = 0

previous_month = 1

for index, x in spy.iterrows():
    current_month = x['month']
    if (current_month != previous_month):
        if(x['MA50_pctchng'] < bear_mean):
            freq_purchase += 1
            number_of_shares += (4000/x['Close'])
            total_cost += 4000
            previous_month = current_month
        elif (x['MA50_pctchng'] > bull_mean):
            freq_purchase += 1
            number_of_shares += (2000/x['Close'])
            total_cost += 2000
            previous_month = current_month
        elif (current_month != int(x['next_month'])):
            freq_purchase += 1
            number_of_shares += (3000/x['Close'])
            total_cost += 3000
            previous_month = current_month
    

end_portfolio_value = number_of_shares * spy.iloc[-1]['Close']
total_returns = end_portfolio_value - total_cost
pct_return = total_returns/total_cost

print('Number of times bought:', freq_purchase)
print('Number of shares:', number_of_shares)
print('Total Cost:', total_cost)
print('End Portfolio Value', end_portfolio_value)
print('Total Returns', total_returns)
print('% Return', pct_return)
```

```python
# DCA 20EMA 2006 - 2019
number_of_shares = 0

freq_purchase = 0

total_cost = 0

previous_month = 1

mean_pct_chng = spy['EMA20_pctchng'].mean()

for index, x in spy.iterrows():
    current_month = x['month']
    if (current_month != previous_month):
        if(x['EMA20_pctchng'] < -mean_pct_chng):
            freq_purchase += 1
            number_of_shares += (4000/x['Close'])
            total_cost += 4000
            previous_month = current_month
        elif (x['EMA20_pctchng'] > mean_pct_chng):
            freq_purchase += 1
            number_of_shares += (2000/x['Close'])
            total_cost += 2000
            previous_month = current_month
        else:
            previous_month = current_month

end_portfolio_value = number_of_shares * spy.iloc[-1]['Close']
total_returns = end_portfolio_value - total_cost
pct_return = total_returns/total_cost

print('Number of times bought', freq_purchase)
print('Number of shares:', number_of_shares)
print('Total Cost:', total_cost)
print('End Portfolio Value', end_portfolio_value)
print('Total Returns', total_returns)
print('% Return', pct_return)
```

```python
import pandas_datareader.data as web
import datetime
pd.set_option('display.width', 200) # Displaying more columns in one row

# Data date range, Google provides up to 4000 entries in one call
start = datetime.datetime(2006, 4, 11) 
end = datetime.datetime(2019, 7, 9)

spy_mod = web.DataReader("SPY", "yahoo", start, end)

value_price = spy_mod['Close'][-1] # The final value of our stock


print(spy_mod.head()) # See first few rows
print(spy_mod.tail()) # See last few rows
```

```python
def doDCA(investment, start_date):
    # Get 12 investment dates in 30 day increments starting from start date
    investment_dates_all = pd.date_range(start_date,periods=159,freq='30D')
    # Remove those dates beyond our known data range
    investment_dates = investment_dates_all[investment_dates_all < spy_mod.index[-1]]

    # Get closest business dates with available data
    closest_investment_dates = spy_mod.index.searchsorted(investment_dates)

    # How much to invest on each date
    portion = investment/159 # (Python 3.0 does implicit double conversion, Python 2.7 does not)

    # Get the total of all stocks purchased for each of those dates (on the Close)
    stocks_invested = sum(portion / spy_mod['Close'][closest_investment_dates])

    # Add uninvested amount back
    uninvested_dollars = portion * sum(investment_dates_all >= spy_mod.index[-1])

    # value of stocks today
    total_value = value_price*stocks_invested + uninvested_dollars
    print('Portion:', portion)
    print('Number of shares:', stocks_invested)
    print('% Return:', (total_value-477000)/477000)
    print('Total Returns:', total_value)
    return total_value

doDCA(477000, spy_mod.index[0])
```
